```python
from itertools import combinations
strng = input().split()
k = int(strng[1])
strn = str(strng[0])
print(k)
print(strn)
sorted_strn = sorted(strn)
print(sorted_strn)
res = [char for char in sorted_strn if char.isupper()]
print(res)

for i in range(0,k+1):
    #print((list(combinations(res, k))))

    per = combinations(res, i)
    for val in per:
        word = ''.join(val)
        print(word)
```
